```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np

# Load the dataset
data = pd.read_csv("nse_all_stock_data (1).csv")

# Check for missing values and drop columns with 100% missing data
data_cleaned = data.dropna(axis=1, how='all')

# Fill missing values using forward fill and backward fill
data_cleaned.fillna(method='ffill', inplace=True)
data_cleaned.fillna(method='bfill', inplace=True)

# Drop any remaining rows with missing values
data_cleaned.dropna(inplace=True)

# Set the 'Date' column as datetime and use it as the index
data_cleaned['Date'] = pd.to_datetime(data_cleaned['Date'])
data_cleaned.set_index('Date', inplace=True)

# Display summary statistics of the cleaned data
summary_stats = data_cleaned.describe()
print(summary_stats)

# Display the first few rows of the cleaned data
print(data_cleaned.head())
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import numpy as np
import matplotlib.pyplot as plt

# Preparing the data for prediction

# Create a new column 'Target' which is the shifted 'RELIANCE' closing price by 1 day (i.e., the price we want to predict)
data_cleaned['Target'] = data_cleaned['RELIANCE'].shift(-1)

# Drop the last row (as it will have NaN in the 'Target' column)
data_cleaned.dropna(inplace=True)

# Define features (X) and target (y)
X = data_cleaned[['RELIANCE']]  # You can add more features here if needed
y = data_cleaned['Target']

# Split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, shuffle=False)

# Initialize the Linear Regression model
model = LinearRegression()

# Train the model
model.fit(X_train, y_train)

# Make predictions on the test set
y_pred = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse}")
print(f"R-squared: {r2}")

# Visualize the actual vs predicted values
plt.figure(figsize=(14, 7))
plt.plot(y_test.index, y_test, label='Actual Price', color='blue', linewidth=2)
plt.plot(y_test.index, y_pred, label='Predicted Price', color='red', linewidth=2)
plt.xlabel('Date')
plt.ylabel('Price')
plt.title('RELIANCE Stock Price Prediction vs Actual (Linear Regression)')
plt.legend()
plt.show()
```
